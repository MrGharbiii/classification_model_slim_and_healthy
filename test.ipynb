# Data Import
Let's import the necessary libraries and load our Excel dataset.

```python
import pandas as pd
import numpy as np

# Load the Excel file
df = pd.read_excel('dataset-ob (1).xlsx')

# Display basic information about the dataset
print(f"Dataset shape: {df.shape}")
print(f"\nColumn names:")
print(df.columns.tolist())
```

```python
valid_profiles = {
    "metabolique": "metabolique",
    "métabolique": "metabolique",
    "meatboalique": "metabolique",
    "metabolique": "metabolique",
    "Metabolique": "metabolique",
    "métabolique": "metabolique",
    "meatboalique": "metabolique",

    "hormonal": "hormonal",
    "h++": "hormonal",
    "H++": "hormonal",
    "h+": "hormonal",
    "H+": "hormonal",
    "h": "hormonal",
    "H": "hormonal",

    "hormonale": "hormonal",

    "digestif": "digestif",

    "psychologique": "psychologique",
    "psy": "psychologique",
    "psy+": "psychologique",
    "psy++": "psychologique",
    "psy+++": "psychologique",
    "pssy": "psychologique",
    "psy": "psychologique",
    "Psy": "psychologique",
    "psy++": "psychologique",
    "Psy++": "psychologique",
    "psy+++": "psychologique",
    "Psy+++": "psychologique",

    "iatrogene": "iatrogene",
    "iatrogène": "iatrogene",
    "itarogéne": "iatrogene",

    "chrono": "iatrogene",
    "chrno": "iatrogene",
    "chrono+": "iatrogene"
}
def clean_label(label_str):
    # Lowercase and remove +
    raw = str(label_str).lower().replace("+", "")
    # Split by "/" and clean
    parts = [p.strip() for p in raw.split("/")]

    # Map to valid profiles
    cleaned = []
    for p in parts:
        standard = valid_profiles.get(p)
        if standard and standard not in cleaned:
            cleaned.append(standard)

    return sorted(cleaned)  # Alphabetical order

df["LABEL_LIST"] = df["LABEL"].apply(clean_label)
print(df)
```

```python
# Examine the values of each feature
print("=== FEATURE VALUES ANALYSIS ===\n")

for feature in feature_columns:
    print(f"📊 Feature: {feature}")
    print(f"   Data type: {X[feature].dtype}")
    print(f"   Missing values: {X[feature].isnull().sum()}/{len(X)} ({X[feature].isnull().sum()/len(X)*100:.1f}%)")
    
    if X[feature].dtype == 'object':
        # For categorical features, show unique values
        unique_vals = X[feature].value_counts().head(100)
        print(f"   Unique values ({X[feature].nunique()} total):")
        for val, count in unique_vals.items():
            print(f"     '{val}': {count}")
    else:
        # For numerical features, show statistics
        print(f"   Statistics:")
        print(f"     Min: {X[feature].min()}")
        print(f"     Max: {X[feature].max()}")
        print(f"     Mean: {X[feature].mean():.2f}" if not X[feature].isnull().all() else "     Mean: N/A")
        print(f"     Non-null values: {X[feature].count()}")
    
    print("-" * 50)
```

```python
# Data Preprocessing
print("=== DATA PREPROCESSING ===\n")

# Create a copy for preprocessing
X_processed = X.copy()

# 1. Handle "NF" (Not Found) values in lab results
lab_columns = ['HOMA-IR', 'A1c', 'VitD', 'TSH', 'Prl', 'Testostérone']
for col in lab_columns:
    if col in X_processed.columns:
        X_processed[col] = X_processed[col].replace('NF', np.nan)
        # Convert to numeric
        X_processed[col] = pd.to_numeric(X_processed[col], errors='coerce')

# 2. Clean categorical variables
# Remove extra spaces
for col in X_processed.select_dtypes(include=['object']).columns:
    X_processed[col] = X_processed[col].astype(str).str.strip()

# 3. Handle "NC" (Not Communicated) values
X_processed = X_processed.replace('NC', np.nan)

# 4. Convert birth date to age
from datetime import datetime
current_year = datetime.now().year
X_processed['Age'] = current_year - pd.to_datetime(X_processed['Date de naissance ']).dt.year
X_processed = X_processed.drop('Date de naissance ', axis=1)

# 5. Convert number of children to numeric
X_processed['Nb enfants'] = pd.to_numeric(X_processed['Nb enfants'], errors='coerce')

print("Preprocessing completed!")
print(f"Dataset shape after preprocessing: {X_processed.shape}")
print(f"Missing values summary:")
print(X_processed.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Detailed analysis of preprocessed data
print("=== PREPROCESSED DATA ANALYSIS ===\n")

for feature in X_processed.columns:
    print(f"📊 Feature: {feature}")
    print(f"   Data type: {X_processed[feature].dtype}")
    print(f"   Missing values: {X_processed[feature].isnull().sum()}/{len(X_processed)} ({X_processed[feature].isnull().sum()/len(X_processed)*100:.1f}%)")
    
    if X_processed[feature].dtype == 'object':
        # For categorical features, show unique values
        unique_vals = X_processed[feature].value_counts().head(10)
        print(f"   Unique values ({X_processed[feature].nunique()} total):")
        for val, count in unique_vals.items():
            print(f"     '{val}': {count}")
    else:
        # For numerical features, show statistics
        print(f"   Statistics:")
        print(f"     Min: {X_processed[feature].min()}")
        print(f"     Max: {X_processed[feature].max()}")
        if not X_processed[feature].isnull().all():
            print(f"     Mean: {X_processed[feature].mean():.2f}")
            print(f"     Std: {X_processed[feature].std():.2f}")
        else:
            print(f"     Mean: N/A")
        print(f"     Non-null values: {X_processed[feature].count()}")
    
    print("-" * 60)

# Summary of changes after preprocessing
print("\n=== PREPROCESSING SUMMARY ===")
print(f"Original shape: {X.shape}")
print(f"Processed shape: {X_processed.shape}")
print(f"\nData types after preprocessing:")
print(X_processed.dtypes.value_counts())
print(f"\nTotal missing values: {X_processed.isnull().sum().sum()}")
print(f"Features with missing values:")
missing_features = X_processed.isnull().sum()
missing_features = missing_features[missing_features > 0].sort_values(ascending=False)
for feature, count in missing_features.items():
    print(f"  {feature}: {count} ({count/len(X_processed)*100:.1f}%)")
```

```python
# Analyze the terrain familial feature
print("=== TERRAIN FAMILIAL ANALYSIS ===")
print(f"Unique values in terrain familial:")
terrain_values = X_processed['terrain familial'].value_counts()
print(terrain_values)

print("\n=== Sample values ===")
for i, val in enumerate(X_processed['terrain familial'].dropna().unique()[:20]):
    print(f"{i+1:2d}. '{val}'")

print(f"\nTotal unique values: {X_processed['terrain familial'].nunique()}")
print(f"Missing values: {X_processed['terrain familial'].isnull().sum()}")
```

```python
# Create binary family history features from terrain familial
print("\n=== CREATING BINARY FAMILY HISTORY FEATURES ===")

# Define the family history conditions to extract
family_conditions = [
    'DT1',           # Diabète type 1
    'DT2',           # Diabète type 2  
    'dysthyroidie',  # Dysthyroïdie
    'HTA',           # Hypertension artérielle
    'obésité',       # Obésité (handling variations)
    'dyslipidémie',  # Dyslipidémie (handling variations)
    'Coronaropathie',# Coronaropathie
    'Autisme',       # Autisme
    'polypose familiale', # Polypose familiale
    'Deficit17alphaOH'    # Déficit 17alpha-OH
]

# Create a working copy of X_processed
X_enhanced = X_processed.copy()

# Function to check if a condition is present in terrain familial
def has_family_history(terrain_value, condition):
    if pd.isna(terrain_value) or terrain_value == 'non':
        return 0
    
    terrain_str = str(terrain_value).lower()
    
    # Handle variations in naming
    condition_variations = {
        'obésité': ['obésité', 'obésite'],  # Handle accent variations
        'dyslipidémie': ['dyslipidémie', 'dyslipidemie', 'dyslipidémie/dyslipidémie'],
        'dysthyroidie': ['dysthyroidie', 'dysthyroïdie']
    }
    
    if condition in condition_variations:
        return int(any(var in terrain_str for var in condition_variations[condition]))
    else:
        return int(condition.lower() in terrain_str)

# Create binary features for each family history condition
for condition in family_conditions:
    feature_name = f'fh_{condition.lower().replace(" ", "_")}'
    X_enhanced[feature_name] = X_enhanced['terrain familial'].apply(
        lambda x: has_family_history(x, condition)
    )
    
    # Count positive cases
    positive_count = X_enhanced[feature_name].sum()
    print(f"{feature_name:20s}: {positive_count:3d} patients ({positive_count/len(X_enhanced)*100:.1f}%)")

# Remove the original terrain familial feature
X_enhanced = X_enhanced.drop('terrain familial', axis=1)

print(f"\nOriginal features: {X_processed.shape[1]}")
print(f"Enhanced features: {X_enhanced.shape[1]}")
print(f"Added binary family history features: {len(family_conditions)}")
```

```python
# Verify the family history features
print("\n=== VERIFICATION OF FAMILY HISTORY FEATURES ===")

# Show some examples of the transformation
print("\nExamples of terrain familial -> binary features transformation:")
for i in [10, 25, 50, 75, 100]:
    original = X_processed.iloc[i]['terrain familial']
    print(f"\nPatient {i}: '{original}'")
    
    # Show which binary features are activated
    fh_cols = [col for col in X_enhanced.columns if col.startswith('fh_')]
    active_features = []
    for col in fh_cols:
        if X_enhanced.iloc[i][col] == 1:
            active_features.append(col)
    
    if active_features:
        print(f"  -> Active: {', '.join(active_features)}")
    else:
        print(f"  -> No family history")

# Summary statistics
print("\n=== FAMILY HISTORY SUMMARY ===")
fh_cols = [col for col in X_enhanced.columns if col.startswith('fh_')]
fh_summary = X_enhanced[fh_cols].sum().sort_values(ascending=False)
print("Family history prevalence:")
for condition, count in fh_summary.items():
    print(f"{condition:20s}: {count:3d} patients ({count/len(X_enhanced)*100:.1f}%)")

print(f"\nPatients with any family history: {(X_enhanced[fh_cols].sum(axis=1) > 0).sum()} ({(X_enhanced[fh_cols].sum(axis=1) > 0).mean()*100:.1f}%)")
print(f"Patients with no family history: {(X_enhanced[fh_cols].sum(axis=1) == 0).sum()} ({(X_enhanced[fh_cols].sum(axis=1) == 0).mean()*100:.1f}%)")
```

```python
# Update X_processed with the enhanced features
X_processed = X_enhanced.copy()
print(f"\n=== FEATURE SET UPDATED ===")
print(f"X_processed now has {X_processed.shape[1]} features")
print(f"Shape: {X_processed.shape}")

# Check for remaining problematic features
print("\n=== CHECKING FOR REMAINING ISSUES ===")

# Check data types
print("\nData types:")
dtype_counts = X_processed.dtypes.value_counts()
print(dtype_counts)

# Check for features with high missing values
print("\nMissing values by feature:")
missing_summary = X_processed.isnull().sum().sort_values(ascending=False)
high_missing = missing_summary[missing_summary > 0]

if len(high_missing) > 0:
    print("Features with missing values:")
    for feature, count in high_missing.items():
        pct = count / len(X_processed) * 100
        print(f"{feature:30s}: {count:3d} ({pct:5.1f}%)")
else:
    print("No missing values found!")

# Check for non-numeric features that might need encoding
print("\nNon-numeric features:")
non_numeric = X_processed.select_dtypes(include=['object']).columns.tolist()
if non_numeric:
    for col in non_numeric:
        unique_vals = X_processed[col].nunique()
        print(f"{col:30s}: {unique_vals} unique values")
        if unique_vals <= 10:  # Show values for categorical features
            print(f"    Values: {list(X_processed[col].unique())}")
else:
    print("All features are numeric!")
```
